````python
import random

game_status = {
    "is_playing": False,
    "current_round": 0,
    "user_points": 0,
    "computer_points": 0,
    "current_winner": None
}

def get_user_choice():
    choices = ["rock", "paper", "scissors"]
    while True:
        user_input = input("Elige: rock, paper o scissors: ").lower()
        if user_input in choices:
            return user_input
        else:
            print("Entrada inválida. Por favor, elige 'rock', 'paper' o 'scissors'.")

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "tie"
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        return "user"
    else:
        return "computer"

def play_game():
    start_game()
    while game_status["is_playing"]:
        play_round()
        show_game_status()
        game_status["is_playing"] = ask_yes_no("¿Deseas seguir jugando? (Y/N): ")
    print("\nGracias por jugar :)")

def start_game():
    game_status["is_playing"] = ask_yes_no("¿Te gustaría jugar? (Y/N): ")
    if game_status["is_playing"]:
        print("\n¡Genial! Comencemos.")
    else:
        print("¡Hasta la próxima!")

def ask_yes_no(prompt):
    while True:
        answer = input(prompt).strip().upper()
        if answer in ('Y', 'N'):
            return answer == 'Y'
        else:
            print("Por favor, responde con 'Y' o 'N'.")

def play_round():
    game_status["current_round"] += 1
    user_choice = get_user_choice()
    computer_choice = get_computer_choice()
    print(f"\nTú elegiste: {user_choice}")
    print(f"La computadora eligió: {computer_choice}")
    winner = determine_winner(user_choice, computer_choice)
    game_status["current_winner"] = winner
    if winner == "user":
        game_status["user_points"] += 1
        print("¡Ganaste esta ronda!")
    elif winner == "computer":
        game_status["computer_points"] += 1
        print("La compu gana.")
    else:
        print("Esta ronda fue un empate.")

def show_game_status():
    print("-" * 30)
    print(f"Ronda actual: {game_status['current_round']}")
    print(f"Puntos del jugador: {game_status['user_points']}")
    print(f"Puntos de la computadora: {game_status['computer_points']}")
    print(f"Ganador de la última ronda: {game_status['current_winner']}")
    print("-" * 30)

if __name__ == "__main__":
    play_game()
````

````python
import random

print("Máquina Tragaperras")

dinero_total_ingresado = 0
dinero_total_ganado = 0
frutas = ["Cerezas", "Naranjas", "Ciruelas", "Campanas", "Melones", "Barras"]

while True:
    dinero = int(input("\nIngresar cantidad de dinero en $COP: "))
    dinero_total_ingresado += dinero

    linea_aleatoria = random.choices(frutas, k=3)
    print("Resultado:", linea_aleatoria)

    if linea_aleatoria[0] == linea_aleatoria[1] == linea_aleatoria[2]:
        ganado = dinero * 3
        print("¡Felicitaciones! Las tres frutas coinciden. Has ganado:", ganado, "COP")
    elif linea_aleatoria[0] == linea_aleatoria[1] or \
         linea_aleatoria[0] == linea_aleatoria[2] or \
         linea_aleatoria[1] == linea_aleatoria[2]:
        ganado = dinero * 2
        print("¡Bien! Dos frutas coinciden. Has ganado:", ganado, "COP")
    else:
        ganado = 0
        print("No hay coincidencias. No has ganado esta vez.")

    dinero_total_ganado += ganado

    jugar_de_nuevo = input("\n¿Deseas jugar de nuevo? (sí/no): ").strip().lower()
    if jugar_de_nuevo != "si":
        break

print("Dinero total ingresado: ", dinero_total_ingresado, "COP")
print("Dinero total ganado: ", dinero_total_ganado, "COP")
````

````python
import random

def jugar_tragaperras():
    frutas = ["Cerezas", "Naranjas", "Ciruelas", "Campanas", "Melones", "Barras"]
    dinero_total_ingresado = 0
    dinero_total_ganado = 0

    while True:
        try:
            dinero = int(input("Ingresar cantidad de dinero en $COP: "))
        except ValueError:
            print("Por favor ingresa un valor numérico.")
            continue

        dinero_total_ingresado += dinero
        linea_aleatoria = [random.choice(frutas) for _ in range(3)]
        print("Resultado:", linea_aleatoria)

        if linea_aleatoria[0] == linea_aleatoria[1] == linea_aleatoria[2]:
            ganado = dinero * 3
            print(f"¡Felicitaciones! Las tres frutas coinciden. Has ganado: {ganado} COP")
        elif linea_aleatoria[0] == linea_aleatoria[1] or linea_aleatoria[0] == linea_aleatoria[2] or linea_aleatoria[1] == linea_aleatoria[2]:
            ganado = dinero * 2
            print(f"¡Bien! Dos frutas coinciden. Has ganado: {ganado} COP")
        else:
            ganado = 0
            print("No hay coincidencias. No has ganado esta vez.")

        dinero_total_ganado += ganado

        jugar_de_nuevo = input("¿Deseas jugar de nuevo? (sí/no): ").lower()
        if jugar_de_nuevo != "sí":
            break

    print(f"Dinero total ingresado: {dinero_total_ingresado} COP")
    print(f"Dinero total ganado: {dinero_total_ganado} COP")

jugar_tragaperras()
````

````python

usuarios = {
    "admin": {"password": "adminpass", "rol": "admin"},
    "juan": {"password": "vendedor123", "rol": "vendedor"}
}

inventario = {
    "001": {"nombre": "Cemento gris x50kg", "cantidad": 100, "precio": 40000},
    "002": {"nombre": "Arena de río m3", "cantidad": 50, "precio": 30000},
    "003": {"nombre": "Grava m3", "cantidad": 75, "precio": 35000},
}

ventas = []

def registrar_usuario():
    nombre = input("Nombre de usuario: ")
    if nombre in usuarios:
        print("El usuario ya existe.")
        return
    password = input("Contraseña: ")
    rol = input("Rol (admin/vendedor): ").lower()
    if rol in ["admin", "vendedor"]:
        usuarios[nombre] = {"password": password, "rol": rol}
        print("Usuario registrado.")
    else:
        print("Rol no válido.")

def agregar_actualizar_producto():
    codigo = input("Código del producto: ")
    nombre = input("Nombre del producto: ")
    cantidad = int(input("Cantidad: "))
    precio = int(input("Precio: "))
    inventario[codigo] = {"nombre": nombre, "cantidad": cantidad, "precio": precio}
    print("Producto agregado o actualizado.")

def consultar_inventario():
    buscar = input("Buscar por código (deja vacío para ver todo): ")
    if buscar:
        producto = inventario.get(buscar)
        if producto:
            print(f"{buscar}: {producto}")
        else:
            print("Producto no encontrado.")
    else:
        for cod, datos in inventario.items():
            print(f"{cod}: {datos}")

def comprar_productos():
    codigo = input("Código del producto: ")
    cantidad = int(input("Cantidad a comprar: "))
    if codigo in inventario:
        inventario[codigo]["cantidad"] += cantidad
    else:
        nombre = input("Nombre del nuevo producto: ")
        precio = int(input("Precio: "))
        inventario[codigo] = {"nombre": nombre, "cantidad": cantidad, "precio": precio}
    print("Compra registrada.")

def ver_resumen_ventas():
    total = 0
    for venta in ventas:
        print(venta)
        total += venta["total"]
    print("Total generado:", total)

def vender_producto():
    codigo = input("Código del producto: ")
    cantidad = int(input("Cantidad a vender: "))
    if codigo in inventario and inventario[codigo]["cantidad"] >= cantidad:
        inventario[codigo]["cantidad"] -= cantidad
        total = inventario[codigo]["precio"] * cantidad
        ventas.append({"codigo": codigo, "nombre": inventario[codigo]["nombre"], "cantidad": cantidad, "total": total})
        print("Venta realizada. Total:", total)
    else:
        print("No hay suficiente stock o el producto no existe.")

def menu_usuario(usuario):
    rol = usuarios[usuario]["rol"]
    while True:
        print("\n--- Menú ---")
        if rol == "admin":
            print("1. Registrar usuario\n2. Agregar/Actualizar producto\n3. Consultar inventario\n4. Comprar productos\n5. Ver resumen de ventas\n6. Cerrar sesión")
            opcion = input("Opcion: ")
            if opcion == "1": registrar_usuario()
            elif opcion == "2": agregar_actualizar_producto()
            elif opcion == "3": consultar_inventario()
            elif opcion == "4": comprar_productos()
            elif opcion == "5": ver_resumen_ventas()
            elif opcion == "6": break
        else:
            print("1. Consultar inventario\n2. Vender producto\n3. Cerrar sesión")
            opcion = input("Opcion: ")
            if opcion == "1": consultar_inventario()
            elif opcion == "2": vender_producto()
            elif opcion == "3": break

def iniciar_sesion():
    usuario = input("Usuario: ")
    contrasena = input("Contraseña: ")
    if usuario in usuarios and usuarios[usuario]["password"] == contrasena:
        print("Acceso permitido")
        menu_usuario(usuario)
    else:
        print("Credenciales incorrectas")

# Inicio del sistema
def main():
    while True:
        print("1. Iniciar sesión\n2. Salir")
        opcion = input("Selecciona una opción: ")
        if opcion == "1": iniciar_sesion()
        elif opcion == "2": break


````
